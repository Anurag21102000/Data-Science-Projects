```python
# Cartoonify Images Through Gaussian Filter 
```

```python
#importing libraries
import cv2
import  numpy as np
```

```python
# reading image
img = cv2.imread("E:\github\Face-X\Cartoonify-Image\Cartoonized-Image\images\smile.jpg")
from skimage import io
io.imshow(img)
```

```python
#converting from BGR2RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
# detecting edges of the img using gaussian-filter
gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 7)
img_edge = cv2.adaptiveThreshold(gray, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 3, 2)
img_edge = cv2.cvtColor(img_edge, cv2.COLOR_GRAY2RGB)
io.imshow(img_edge)
```

```python
# cartoonifying image using bilateral filter technique
color = cv2.bilateralFilter(img,15,30,20)
cartoon = cv2.bitwise_and(color, img_edge)
io.imshow(cartoon)
```
